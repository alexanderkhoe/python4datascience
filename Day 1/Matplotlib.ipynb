```python
%matplotlib inline
import numpy as np
from matplotlib.pyplot import * 

x = np.arange(1, 5) 
plot(x, x*1.5, label='Normal') 
plot(x, x*3.0, label='Fast') 
plot(x, x/3.0, label='Slow') 
grid(True) 
title('Sample Growth of a Measure') 
xlabel('Samples') 
ylabel('Values Measured') 
legend(loc='upper left')
savefig('plot123.png') #save to file in current directory
show()
```

```python
%matplotlib inline
from pylab import *

data1 = 10*rand(5) 
data2 = 10*rand(5) 
data3 = 10*rand(5) 
e2 = 0.5 * abs(randn(len(data2))) 
locs = arange(1, len(data1)+1) 
width = 0.27 
bar(locs, data1, width=width);
bar(locs+width, data2, yerr=e2, width=width,color='red');
bar(locs+2*width, data3, width=width, color='green') ;
xticks(locs + width*1.5, locs);
show()
```

```python
%matplotlib inline
from pylab import *

axis([0, 10, 0, 20]);
arrstyles = ['-', '->', '-[', '<-', '<->', 'fancy', 'simple', 
'wedge'] 
for i, style in enumerate(arrstyles):
    annotate(style, xytext=(1, 2+2*i), xy=(4, 1+2*i), arrowprops=dict(arrowstyle=style));
    connstyles=["arc", "arc,angleA=10,armA=30,rad=15", "arc3,rad=.2", "arc3,rad=-.2", "angle", "angle3"]
    for i, style in enumerate(connstyles):
        annotate("", xytext=(6, 2+2*i), xy=(8, 1+2*i), arrowprops=dict(arrowstyle='->', connectionstyle=style));
show()
```

```python
%matplotlib notebook

def func3(x,y):
    return (1- x/2 + x**5 + y**3)*exp(-x**2-y**2)


dx, dy = 0.05, 0.05
x = arange(-3.0, 3.0001, dx)
y = arange(-3.0, 3.0001, dy)
X,Y = meshgrid(x, y)
Z = func3(X, Y)
pcolor(X, Y, Z)
colorbar()
axis([-3,3,-3,3])
show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D 

import matplotlib.pyplot as plt
import numpy as np


x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(x, y)

# Compute z to make the surface.
z = X**2 + Y**2

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(X, Y, z)

plt.show()
```
