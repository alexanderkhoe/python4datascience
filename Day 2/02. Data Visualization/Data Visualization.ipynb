# Data Visualization

```python
# Plot 1: List, Matplotlib
%matplotlib notebook

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

import matplotlib.pyplot as plt

plt.plot(year,pop,'g--')
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# Plot 2: Array, Matplotlib, Subplot
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.1)

plt.subplot(211)
plt.plot(x, np.sin(x), 'y')
plt.title("Sinus")

plt.subplot(212)
plt.plot(x, np.cos(x), 'o')
plt.title("Cosinus")

plt.show()

```

```python
# Plot 3: DataFrame, Matplotlib, Subplot, Histogram
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd

tips = pd.read_csv("tips.csv")
tip = tips["tip"]
bill = tips["total_bill"]
total = bill+tip

plt.subplot(121)
plt.hist(tip, color = "skyblue")
plt.title("Tip")
plt.axis('off')

plt.subplot(122)
plt.hist(total,color = "red")
plt.title("Total")
plt.axis('equal')

plt.show()
```

```python
# Plot 4: Array, Meshgrid, Matplotlib 2D
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

u = np.arange(-5,5,0.5)
v = np.arange(-5,5,0.5)

X,Y = np.meshgrid(u,v)
Z = X**2 - Y**2

plt.set_cmap("winter")
plt.pcolor(X,Y,Z)
plt.show()


```

```python
# Plot 5: DataFrame, Seaborn
%matplotlib notebook

import matplotlib.pyplot as plt
import seaborn as sb

tips = sb.load_dataset("tips")
tips.head()

#sb.lmplot('total_bill', 'tip', data=tips)
sb.lmplot('total_bill', 'tip', data=tips, hue='sex')
#sb.lmplot('total_bill', 'tip', data=tips, col='sex')

plt.show()
```

```python
# Plot 6: DataFrame, Seaborn, Combine
%matplotlib notebook

import matplotlib.pyplot as plt
import seaborn as sb

tips = sb.load_dataset("tips")

sb.boxplot(x='day', y='tip', data=tips)
sb.violinplot(x='day', y='tip', data=tips, color='lightgray')
sb.stripplot(x='day', y='tip', data=tips)


plt.ylabel('tip ($)')
plt.show()
```
